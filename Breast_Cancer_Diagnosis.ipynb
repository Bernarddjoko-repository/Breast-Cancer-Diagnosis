```python
### Standardize Numerical Features
```

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('Diagnosis', axis=1)
y = df['Diagnosis']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame (for feature names)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

X_scaled_df.head()
```

```python
# Create a new feature: This might capture density or shape variation
X_scaled_df['area_ratio'] = df['area_mean'] / df['perimeter_mean']
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Fit model to get feature importance
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_scaled_df, y)

# Create importance dataframe
importances = pd.DataFrame({
    'feature': X_scaled_df.columns,
    'importance': rf.feature_importances_
}).sort_values(by='importance', ascending=False)

# Display top 10 important features
importances.head(10)
```

```python
# Visualize top features
top_features = importances.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_features, x='importance', y='feature', palette='viridis')
plt.title('Top 10 Feature Importances (Random Forest)')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.show()
```

Model Developement

```python
# Select top 10 features from the importance list
top_10_features = importances.head(10)['feature'].tolist()

# Create reduced feature set
X_top10 = X_scaled_df[top_10_features]
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import pandas as pd

# Step 1: Split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X_top10, y, test_size=0.2, stratify=y, random_state=42)

# Step 2: Apply SMOTE to training data only
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check class distribution
print("Original Training Class Distribution:\n", y_train.value_counts())
print("\nResampled Training Class Distribution:\n", pd.Series(y_train_resampled).value_counts())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert resampled labels to a DataFrame
y_resampled_df = pd.DataFrame(y_train_resampled, columns=['Diagnosis'])
y_resampled_df['Diagnosis'] = y_resampled_df['Diagnosis'].map({0: 'Benign', 1: 'Malignant'})

# Plot
plt.figure(figsize=(5, 4))
sns.countplot(x='Diagnosis', data=y_resampled_df, palette='pastel')
plt.title("Class Distribution After SMOTE")
plt.ylabel("Number of Samples")
plt.xlabel("Diagnosis")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Define models
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'SVM': SVC(kernel='rbf', probability=True, random_state=42)
}

# Train on resampled training data
for name, model in models.items():
    model.fit(X_train_resampled, y_train_resampled)
    print(f"{name} trained on SMOTE-balanced data.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Evaluate performance
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"\nðŸ“Š Evaluation for: {name}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate predictions
y_pred_dt = models['Decision Tree'].predict(X_test)
y_pred_knn = models['KNN'].predict(X_test)
y_pred_svm = models['SVM'].predict(X_test)

# Set up subplot grid
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Decision Tree
cm_dt = confusion_matrix(y_test, y_pred_dt)
ConfusionMatrixDisplay(cm_dt, display_labels=['Benign', 'Malignant']).plot(ax=axes[0], cmap='Blues')
axes[0].set_title("Decision Tree")

# KNN
cm_knn = confusion_matrix(y_test, y_pred_knn)
ConfusionMatrixDisplay(cm_knn, display_labels=['Benign', 'Malignant']).plot(ax=axes[1], cmap='Greens')
axes[1].set_title("KNN")

# SVM
cm_svm = confusion_matrix(y_test, y_pred_svm)
ConfusionMatrixDisplay(cm_svm, display_labels=['Benign', 'Malignant']).plot(ax=axes[2], cmap='Oranges')
axes[2].set_title("SVM")

# Adjust layout
plt.suptitle("Confusion Matrices for All Models", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
import numpy as np

# Use 5-fold stratified cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Store results
cv_results = {}

# Evaluate each model using cross_val_score on the *full* dataset (X_top10)
for name, model in models.items():
    scores = cross_val_score(model, X_top10, y, cv=cv, scoring='accuracy')
    cv_results[name] = scores
    print(f"\nâœ… {name} Cross-Validation Results:")
    print("Scores:", scores)
    print("Mean Accuracy:", scores.mean())
    print("Standard Deviation:", scores.std())
```

```python
import matplotlib.pyplot as plt

# Prepare data
models_list = list(cv_results.keys())
means = [cv_results[m].mean() for m in models_list]
stds = [cv_results[m].std() for m in models_list]

# Plot
plt.figure(figsize=(8, 5))
plt.bar(models_list, means, yerr=stds, capsize=5, color=['skyblue', 'lightgreen', 'salmon'])
plt.ylabel('Mean Accuracy')
plt.title('Cross-Validation Accuracy with Error Bars (Â±1 std)')
plt.ylim(0.85, 1.0)
plt.grid(axis='y')
plt.show()
```

**Decision Boundary Visualizations**

```python
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap
import numpy as np

# Reduce to 2D using PCA for plotting
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_top10)

# Create a dictionary of models already trained
trained_models = {
    'Decision Tree': models['Decision Tree'],
    'KNN': models['KNN'],
    'SVM': models['SVM']
}

# Retrain on PCA-reduced data
for name in trained_models:
    trained_models[name].fit(X_pca, y)

# Set up mesh grid for decision surface
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))

# Plot decision boundaries
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])
cmap_bold = ['#FF0000', '#00FF00']

for idx, (name, model) in enumerate(trained_models.items()):
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    axes[idx].contourf(xx, yy, Z, cmap=cmap_light, alpha=0.6)
    axes[idx].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap=ListedColormap(cmap_bold), edgecolor='k', s=20)
    axes[idx].set_title(f"{name} Decision Boundary (PCA 2D)")
    axes[idx].set_xlabel("PCA Component 1")
    axes[idx].set_ylabel("PCA Component 2")

plt.tight_layout()
plt.show()
```
